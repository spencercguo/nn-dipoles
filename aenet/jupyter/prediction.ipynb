```python
dipole_file = '../../data/molecular_dipoles.dat'
```

```python
true_dipoles = read_molecular_dipoles(dipole_file, 200)
```

```python
dipoles.shape
```

```python
prediction_data_z = np.load('../../data/predconfigs-z.p', allow_pickle=True)
prediction_data_x = np.load('../../data/predconfigs-x.p', allow_pickle=True)
prediction_data_y = np.load('../../data/predconfigs-y.p', allow_pickle=True)
```

```python
ndipoles = len(prediction_data_z.keys())
```

```python
pred_dipoles = np.zeros((ndipoles, 3), dtype=np.float32)
for i in range(ndipoles):
    pred_dipoles[i,0] = prediction_data_x[i+1]['energy']
    pred_dipoles[i,1] = prediction_data_y[i+1]['energy']
    pred_dipoles[i,2] = prediction_data_z[i+1]['energy']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(abs(pred_dipoles[:,2]), bins=20)
```

```python
plt.hist(pred_dipoles[:,0], bins=20)
```

```python
plt.hist(pred_dipoles[:,1], bins=20)
```

```python
dipole_mag = np.linalg.norm(pred_dipoles, axis=1)
```

```python
plt.hist(dipole_mag, bins=20)
plt.title('Histogram of NN-predicted dipoles')
plt.xlabel('Dipole moment (D)')
plt.ylabel('Count')
plt.savefig('../fig/nn-dipole-hist', dpi=200)
```

```python
true_mag = np.linalg.norm(true_dipoles, axis=1)
true_mag = true_mag.reshape(25600)
```

```python
plt.hist(true_mag, bins=20)
plt.title('Histogram of DFT-calculated dipoles')
plt.xlabel('Dipole moment (D)')
plt.ylabel('Count')
plt.savefig('../fig/true-dipole-hist', dpi=200)
```

```python
from scipy.stats import linregress
lin = linregress(true_mag, dipole_mag)
m, b = lin[0], lin[1]
print('R^2 value:', lin[2])
```

```python
plt.scatter(true_mag, dipole_mag)
# x = np.unique(true_mag)
# plt.plot(x, b + m * x, 'r')
plt.title('Correlation between DFT and NN dipoles')
plt.xlabel('DFT dipole moment (D)')
plt.ylabel('NN-predicted dipole moment (D)')
plt.savefig('../fig/correlation', dpi=200)
```
